```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def y_function(x):
    return x**2
```

```python
x = range(-100, 101)
y = map(y_function, x)
```

```python
x, y
```

```python
x = list(x)
y = list(y)
x, y
```

```python
plt.plot(x, y)
plt.show()
```

```python
def y_derivative(x):
    return 2*x
```

```python
current_pos = 50, y_function(50)
```

```python
plt.plot(x, y)
plt.scatter(current_pos[0], current_pos[1], color='red')
plt.show()
```

```python
learning_rate = 0.01
```

```python
from IPython.display import clear_output
```

```python
for _ in range(200):
    new_x = current_pos[0] - learning_rate * y_derivative(current_pos[0])
    new_y = y_function(new_x)
    current_pos = new_x, new_y

    plt.plot(x, y)
    plt.scatter(current_pos[0], current_pos[1], color='red')
    plt.pause(0.001)
    # plt.clf()
    plt.close()
    clear_output(wait=True)
    plt.show()
```
